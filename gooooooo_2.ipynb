Data augmentation

```python
import random

def augment_image(image):
  # used to flip the images
    if random.random() > 0.5:
        image = cv2.flip(image, 1)

    h, w = image.shape[:2]


    rotation_matix = cv2.getRotationMatrix2D((w//2, h//2), 45, 1)
    rotated_image = cv2.warpAffine(image, rotation_matix, (w, h)) # rotates the image

    return rotated_image
```

#2.Load Data.

```python

def load_data(dir_list, image_size=(128,128),augment_no_tumor=True):
    X = []
    y = []
    image_width, image_height = image_size


    for directory in dir_list:
        for label_folder in ["tumor", "no_tumor"]:
            folder_path = os.path.join(directory, label_folder)
            for filename in os.listdir(folder_path):
                img_path = os.path.join(folder_path, filename) # finds the file that is stored

                # preprocssing
                image = cv2.imread(img_path)
                image = cv2.resize(image,(image_width, image_height),interpolation=cv2.INTER_CUBIC)
                image = image / 255.0


                X.append(image)
                #label image 1 or 0 deping on tumours

                label = 1 if label_folder == "tumor" else 0
                y.append([label])

                if augment_no_tumor and label == 0:
                  # augments only no tumor images

                    X.append(augment_image(image))
                    y.append([label])

    X = np.array(X)
    y = np.array(y)

    X, y = shuffle(X, y) # shuffles dataset


    return X, y

```

```python
trainingSet = ["brain_tumor_data/Training_binary"]
testSet = ["brain_tumor_data/Testing_binary"]


X_train, y_train = load_data(trainingSet, image_size=(128,128), augment_no_tumor=True)
X_test, y_test = load_data(testSet, image_size=(128,128), augment_no_tumor=False)  # used to load the data

# shows the images amount
print("Training set:")
print(" Tumour images:", np.sum(y_train == 1))
print(" No-tumour images:", np.sum(y_train == 0))

print("\nTest set:")
print(" Tumour images:", np.sum(y_test == 1))
print(" No-tumour images:", np.sum(y_test == 0))


# this is to split data for validation
X_train, X_val, y_train, y_val = train_test_split( X_train,y_train, random_state=42, test_size=0.2, stratify=y_train,)



```

# 3.Plot images

```python

def plot_sample_images(X, y, n=30):
    X = np.array(X)
    y = np.array(y)

    for label in [0, 1]:
      # goes through each image and sees if labeled 1 or 0

        images = np.where(y.flatten() == label)[0]
        n_images = X[images[:n]]
        # used for layout
        columns_n = 10
        rows_n = int(n / columns_n)

        plt.figure(figsize=(20, 4))
        i = 1
        # loops and plots image
        for image in n_images:
            plt.subplot(rows_n, columns_n, i)
            plt.imshow(image)
            plt.axis('off')
            i += 1

        title = "Tumor" if label == 1 else "No Tumor"
        plt.suptitle(title, fontsize=20)
        plt.show()

```

```python

plot_sample_images(X_train, y_train, n=30)
```

# 4.Creating convulution neural network

```python

model = Sequential([
    Conv2D(16, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D((4,4)),
    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D((4,4)),
    Conv2D(64, (3,3), activation='relu'),
    AveragePooling2D((2,2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid') # used as this is binary 1 or 0
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy']) # used as this is binary 1 or 0
model.summary()
```

# 5.Train the model

```python
epochs = 15
batch_size = 64

history = model.fit(
    X_train, y_train, # training set
    validation_data=(X_val, y_val), # validation set
    epochs=epochs,
    batch_size=batch_size
)
```

# 6.output and visualisation

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

# 7.Evaluation

```python
Model_loss, Model_accuracy = model.evaluate(X_test, y_test) # test set is then evalutated

print(f"the accuracy of the model is: {Model_accuracy}")
print(f"the loss of the model is: {Model_loss}")
```

```python

def plot_confusion_matrix(actual, predicted, labels):
    cm = tf.math.confusion_matrix(actual, predicted).numpy()

    plt.figure(figsize=(10, 10))
    ax = sns.heatmap(cm, annot=True, fmt='g')

    ax.set_title('Confusion Matrix')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('actual')
    ax.set_xticklabels(labels, rotation=90)
    ax.set_yticklabels(labels, rotation=0)

    plt.show()
class_names = ['No Tumour', 'Tumour']
predictions = (model.predict(X_test) > 0.5).astype(int).ravel() # done as

plot_confusion_matrix(y_test, predictions, class_names)

```

```python

print(classification_report(
    y_test,
    predictions,
    target_names=['No Tumour', 'Tumour']
))
```

Training the CNN I had made 15 epochs which means that it will go through all images 15 different times during each epoch and I used a batch size of 64 which is for smaller data sets this means that in each epoch it will update the weights used after every 64 images.For me this lead to an average of a minute to process for each epoch using ~7GB of RAM as the more batches cause higher amounts of RAM being used. In the dataset that I used there are 5000 JPEG images that are split 900 into the testing data and 4100 in the training having a total of 158.6MB. To decrease the effects of overfitting and the imbalance it would need to use more images from a larger dataset however to do this there isn't enough RAM that is available, this was shown to me from attempting to increase the dataset however it causes google collab to crash. This shows that to have accuracy within a model there will be massive data storage requirements that are needed.
While saying that when used in production deployment there will be a massive usage of energy that will come from storing these large models and continuously using them for each prediction there will also be continuously retraining the model and constantly updating the model which will use more and more storage and increase the energy that is used. In doing so this will have a massive impact on the environment which will need to be reduced.
There are many ways to reduce environmental impact in production deployment. One of which is optimizing training because the more training the CNN model does the more electricity is used so by optimizing this it will reduce this factor. This can be done by many ways pre-processing data before training and adjusting the epochs and batches used, reducing the training time will improve the usage of electricity drastically. There is also early stopping which can be used to reduce environmental impact as this stops over training as it looks at the validation performance in each epoch and will stop training if there are no improvements. Both of these techniques, optimization and early stoppage, are used to reduce training which reduces energy consumption overall having a reduction in the environmental impact.
